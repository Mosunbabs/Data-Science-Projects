```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

titanic= pd.read_csv("C:\\Users\\USER\\Downloads\\Week One Learning Task\\QUESTION 3\\Titanic-Dataset.csv",index_col=0)
print(titanic.head())
```

```python
print(titanic.info())
```

```python
#checking and finding the sum of missing values per column
print(titanic.isnull().sum())
```

```python
# Filling the missing values in 'Age' with the median
titanic['Age'].fillna(titanic['Age'].median(),inplace=True)

# Filling the missing values in 'Embarked' with the mode
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)

#droping the cabin colunm
titanic.drop(columns=['Cabin'], inplace=True)
print(titanic.isnull().sum())
```

```python
print(titanic.info())
```

```python
#visualization for all numerical values
plt.hist(titanic.Age)
plt.xlabel('Age')
plt.title('Age distribution of passangers')
plt.ylabel('Number of passangers')


plt.show()

```

```python
#bar chart showing the survival rate for each passenger class
sns.barplot(x= titanic.Pclass,y=titanic.Survived)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
#ploting the age distribustion of survivors and non survivors by sex
sns.boxplot(x =titanic.Sex,y = titanic.Age,hue = titanic.Survived)
plt.title("Age Distribution of Survivors and Non-Survivors by Sex")
plt.show()
```

```python
#Exploring the relationship between age and fare Rate by sex
sns.scatterplot(x = titanic.Age,y = titanic.Fare,hue = titanic.Sex)
plt.title("Fare vs. Survival Rate by Sex")
plt.show()
```

```python
# Adding an horizontal line at the average fare
plt.axhline(titanic['Fare'].mean(), color='red', linestyle='--', label='Average Fare')
plt.hist(titanic["Fare"], bins=5)
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution of Passengers")
plt.legend()
plt.show()
```
